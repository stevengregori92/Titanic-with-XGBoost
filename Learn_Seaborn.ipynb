```python
!pip install --upgrade seaborn
import seaborn as sns
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.scatterplot(x = "tip", y="total_bill", data=tips, hue = 'day', size = 'size', palette = 'kjk')
```

```python

sns.displot(tips['tip'], kde = True, bins = 30)
```

```python
sns.barplot(x = "sex", y="tip", data=tips, palette = 'YlGnBu')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', palette = 'YlGnBu', dodge = True)
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = tips, kind = 'reg')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = tips, kind = 'kde', cmap = 'YlGnBu')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = tips, palette = 'YlGnBu', kind = 'kde', shade = True)
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = tips, kind = 'hex', cmap = 'YlGnBu')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.pairplot(titanic.select_dtypes(['number']), hue = 'pclass')
```

```python
sns.heatmap(titanic.corr(), annot = True, cmap = 'BuPu')
```
